```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from konlpy.tag import Okt
import pandas as pd
import numpy as np
from konlpy.tag import Mecab
mecab = Mecab()
from konlpy.tag import Okt
okt = Okt()
```

```python
df_train = pd.read_csv('./POS_ITEM3_CJ.csv', error_bad_lines=False)
```

```python
df_train[:10]
```

```python
df_train['NM'] = df_train['NM'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")#한글과 공백 제외(^ == not)
df_train[:10]
```

```python
df_train['NM'].replace('', np.nan, inplace=True) #위 정제후 빈칸이 되었으면 null(np.nan numpy의 null형태)로 replace
print(len(df_train))
print(df_train.isnull().sum())
df_train = df_train.dropna(how='any')
print(len(df_train))
```

```python
def spacing_okt(wrongSentence):
    tagged = okt.pos(wrongSentence)
    corrected = ""
    for i in tagged:
        if i[1] in ('Josa', 'PreEomi', 'Eomi', 'Suffix', 'Punctuation'):
            corrected += i[0]
        else:
            corrected += " "+i[0]
    if corrected[0] == " ":
        corrected = corrected[1:]
    return corrected
```

```python
dict = {}
for i in range(len(df_train)):
    try:
      if df_train.loc[i, 'CD'] not in dict.keys():
        dict[df_train.loc[i, 'CD']] = ""  
      dict[df_train.loc[i, 'CD']] = dict[df_train.loc[i, 'CD']]+" "+spacing_okt(df_train.loc[i, 'NM'])
      print("Make Dictionary {}".format(i))
    except:
      pass
```

```python
data = []
labels = []
i = 0
for key, val in dict.items():
    #dict[key] = spacing_okt(val)
    print(key)
    print(val)
    data.append(val)
    labels.append(key)
```

```python
tagged_data = [TaggedDocument(words=okt.morphs(_d), tags=[str(i)]) for i, _d in enumerate(data)]
for i in tagged_data:
    print(i)
```

```python
# 5: 378
# 3 : 116
len(data)
```

```python
# 모델 설계
max_epochs = 1000
# https://roboreport.co.kr/doc2vec-%ED%9B%88%EB%A0%A8-%ED%8C%8C%EB%9D%BC%EB%AF%B8%ED%84%B0-%EC%84%A4%EB%AA%85/
model = Doc2Vec(vector_size= 300,
                alpha=0.025, #learning rate
                min_alpha=0.00025,
                min_count=1, #데이터에서 등장하는 단어의 최소빈도수
                dm=1)

#build_vocab
model.build_vocab(tagged_data)

# 학습
for epoch in range(max_epochs):
    print("@@@@ Epoch count : {} @@@@".format(epoch))
    model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=model.iter)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha
```

```python
# save model (classdepth_epoch_vectorsize)
model.save("d2v_3_1000_300_CJ.model")
```

```python
sample = spacing_okt("쁘띠첼밀감110g")
print(sample)
print(model.infer_vector(sample))
```

```python
cor = 0
cnt = 0
cSim = 0
nSim = 0
for i in range(0,df_train.shape[0]):
  # print('Calc similar : {}th'.format(i))
  try:
    sample = spacing_okt(df_train.loc[i, 'NM'])
    vector = model.infer_vector(sample)
    similar_docs = model.docvecs.most_similar([vector], topn=1)
    #if similar_docs[0][1] >= 0.65: #임시 정확도 65퍼만 뽑아본다
    if df_train.loc[i, 'CD']==labels[int(similar_docs[0][0])]:
      #print(similar_docs[0][1])
      print(sample)
      cSim +=  similar_docs[0][1]
      cor += 1;
    else:
      nSim +=  similar_docs[0][1]
    cnt+=1
  except:
    pass
```
